# Read Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.pandas.set_option('display.max_columns', None)

data = pd.read_csv("train.csv")
data.head()
```

# Read data_description file

```python
data_desc = []

with open('data_description.txt', 'r') as f:
    lines = f.readlines()

current_feature = ""
for line in lines:
    # Check if this line starts a new feature description
    if ':' in line:
        if current_feature:
            # Add the previous feature description to the list
            data_desc.append(current_feature.strip())
        # Start a new feature description
        current_feature = line
    else:
        # Continue the current feature description
        current_feature += line

# Add the last feature description to the list
if current_feature:
    data_desc.append(current_feature.strip())


data_desc_dict = {}
def get_feature_desc_in_dict(feature_desc, data_desc_dict, data_desc):
    line = feature_desc.splitlines()[0]
    key = line.split(':')[0].strip()
    
    for line in feature_desc.splitlines():
        if ':' in line:
            data_desc_dict[key] = []
        else:
            if len(line.split()):
                data_desc_dict[key].append(line.split()[0])
            
for feature_desc in data_desc:
    get_feature_desc_in_dict(feature_desc, data_desc_dict, data_desc)

categorical_features = [feature for feature in data.columns if data[feature].dtypes=='O']
for key in categorical_features:
    print(key ,' : ',data_desc_dict[key])
```

# Pipeline Functions

```python
def impute_cat_features(real_data):
    data = real_data.copy()
    null_cat_features = [feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes=='O']
    for feature in null_cat_features:
        data[feature].fillna('Missing', inplace=True)
    return data

def impute_num_features(real_data):
    data = real_data.copy()
    null_num_features = [feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes!='O']
    for feature in null_num_features:
        # data[feature+'Nan'] = np.where(data[feature].isnull(), 1, 0)
        data[feature].fillna(data[feature].median(), inplace=True)
    return data

def handle_year_features(real_data):
    data = real_data.copy()
    year_features = [feature for feature in data.columns if 'Year' in feature or 'Yr' in feature and 'Nan' not in feature]
    for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
        data[feature] = data['YrSold'] - data[feature]
    data.drop('YrSold', axis=1, inplace=True)
    return data

def handle_continous_features(real_data):
    data = real_data.copy()
    numerical_features = [feature for feature in data.columns if data[feature].dtypes!='O']
    year_features = [feature for feature in data.columns if 'Year' in feature or 'Yr' in feature and 'Nan' not in feature]
    continous_features = [feature for feature in numerical_features if len(data[feature].unique())>=25 and feature not in year_features and feature!='Id' and feature!='SalePrice' ]
    for feature in continous_features:
        if 0 in data[feature].unique():
            pass
        else:
            data[feature] = np.log(data[feature])
    return data

def handle_categorical_features(data):
    new_data = data.copy()
    categorical_features = [feature for feature in new_data.columns if new_data[feature].dtypes=='O']
    
    for feature in categorical_features:
        if 'Missing' in new_data[feature].unique():
            dummy_df = pd.get_dummies(new_data[feature], dtype=int, prefix=feature, prefix_sep = '_').drop(f'{feature}_Missing', axis=1)
        else:
            dummy_df = pd.get_dummies(new_data[feature], dtype=int, prefix=feature, prefix_sep = '_')
        
        new_data = pd.concat([new_data, dummy_df], axis=1)
        new_data.drop(feature, axis=1, inplace=True)
    return new_data

# def handle_rare_variables(real_data):
#     data = real_data.copy()
#     categorical_features = [feature for feature in data.columns if data[feature].dtypes=='O']
    
#     for feature in categorical_features:
#         temp = data.groupby(feature)['Id'].count()/len(data)
#         temp_df = temp[temp>0.01].index
#         data[feature] = np.where(data[feature].isin(temp_df), data[feature], 'Rare_Var')
#     return data

def pipeline(real_data):
    data = impute_cat_features(real_data)
    data = impute_num_features(real_data)
    data = handle_year_features(real_data)
    data = handle_continous_features(real_data)
    data = handle_categorical_features(real_data)
    # data = handle_rare_variables(real_data)
    
    return data
    
```

# Running Pipeline

```python
data = pipeline(data)
```
