```python
data_desc = []

with open('data_description.txt', 'r') as f:
    lines = f.readlines()

current_feature = ""
for line in lines:
    # Check if this line starts a new feature description
    if ':' in line:
        if current_feature:
            # Add the previous feature description to the list
            data_desc.append(current_feature.strip())
        # Start a new feature description
        current_feature = line
    else:
        # Continue the current feature description
        current_feature += line

# Add the last feature description to the list
if current_feature:
    data_desc.append(current_feature.strip())


```

```python
# Print the results to check

for feature in data_desc:
    print(feature)
    print("----")

# Verify the length
print(len(data_desc))
```

```python
data_desc_dict = {}
def get_feature_desc_in_dict(feature_desc, data_desc_dict, data_desc):
    line = feature_desc.splitlines()[0]
    key = line.split(':')[0].strip()
    
    for line in feature_desc.splitlines():
        if ':' in line:
            data_desc_dict[key] = []
        else:
            if len(line.split()):
                data_desc_dict[key].append(line.split()[0])
            
for feature_desc in data_desc:
    get_feature_desc_in_dict(feature_desc, data_desc_dict, data_desc)
```
